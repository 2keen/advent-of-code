```python
def bti(b):
    num = 0
    for i in range(len(b)):
        if b[-1::-1][i]=='1':
            num+=2**i
    return num
```

```python
def part1(instructions):
    mem = {}
    for ins in instructions:
        print(ins)
        if ins[0] == 'mask':
            print(ins[1])
            mask = ins[1]
        else:
            n = list('0' * (2+len(mask)-len(bin(int(ins[1])))) + bin(int(ins[1]))[2:])
            for i in range(len(mask)):
                if mask[i]!='X':
                    n[i] = mask[i]
            mem[ins[0][4:-1]] = bti(''.join(n))
            print(ins[0][4:-1], bti(''.join(n)))

    return sum(mem.values())   
```

```python
def bti2(addr):
    addresses = [0]
    for i in range(len(addr)):
        if addr[len(addr)-i-1] == '0': continue
        elif addr[len(addr)-i-1] == '1':
            addresses = [addresses[j] + 2**i for j in range(len(addresses))]

        elif addr[len(addr)-i-1] == 'X':
            addresses = addresses + [addresses[j] + 2**i for j in range(len(addresses))]
        else: print('fail')
        
        print(i,addr[len(addr)-i-1])
    return addresses
```

```python
def bti3(addr, addresses = [0]):
    for i in range(len(addr)):
        if addr[len(addr)-i-1] == '0': continue
        elif addr[len(addr)-i-1] == '1':
            addresses = [addresses[j] + 2**i for j in range(len(addresses))]
        elif addr[len(addr)-i-1] == 'X':
            return addresses+[addresses[j] + 2**i for j in range(len(addresses))]
    return addresses
```

```python

import sys

def prep_mask(linha):
    pX = []
    m1 = ['0'] * 36

    for i, c in enumerate(linha):
        if c == '1':
            m1[i] = '1'
        elif c == 'X':
            pX.append(i)

    return pX, int(''.join(m1), 2)

def parse(linha):
    pre, valor = linha.split(' = ')
    p1, p2 = pre.find('[') + 1, pre.find(']')
    addr = pre[p1:p2]
    return int(addr), int(valor)

def set_bit(value, bit):
    return value | (1<<bit)

def clear_bit(value, bit):
    return value & ~(1<<bit)

pX, m1 = None, None
mem = dict()

def mark(addr, valor, p):
    if p >= len(pX):
        mem[addr] = valor
        return

    mark(set_bit(addr, 35-pX[p]), valor, p+1)
    mark(clear_bit(addr, 35-pX[p]), valor, p+1)

for linha in sys.stdin:
    linha = linha.rstrip()
    if linha[1] == 'a':
        pX, m1 = prep_mask(linha[7:])
    else:
        addr, valor = parse(linha)
        mark(addr | m1, valor, 0)

total = 0
for v in mem.values():
    total += v

print(total)
```
