```python
instructions = open('input.txt').read().split('\n')
instructions[:5]
```

```python
test = ['F10','N3','F7','R90','F11']
```

```python
dmap = {
    'N': (1,0),
    'S': (-1,0),
    'E': (0,1),
    'W': (0,-1),
    0  : 'E',
    180: 'W',
    90 : 'N',
    270: 'S',
}
```

```python

ship = [0,[0,0]]
for t in instructions:
    d = t[0]
    mag = int(t[1:])
    if d == 'L':
        ship[0] = (ship[0] + mag) % 360
    elif d == 'R':
        ship[0] = (ship[0] - mag) % 360
    else:
        if d == 'F': d = dmap[ship[0]]        
        ship[1][0]+= dmap[d][0]*mag
        ship[1][1]+= dmap[d][1]*mag
    
    
print(ship,abs(ship[1][0])+abs(ship[1][1]))   
```
