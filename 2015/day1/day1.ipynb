```python
with open('input.txt') as f:
    directions = f.read()

len(directions)
```

```python
for d in directions:
    print(d)
```

```python
# part 1
floor = 0
for d in directions:
    if d == '(':
        floor += 1 #floor = floor + 1
    elif d == ')':
        floor -= 1
    else:
        print('invalid character {}'.format(d))
print('final floor: {}'.format(floor))
```

```python
# part 2
floor = 0
for i,d in enumerate(directions):
    if d == '(':
        floor += 1
    elif d == ')':
        floor -= 1
    else:
        print('invalid character {}'.format(d))
    if floor < 0:
        print('enter basemet at step: {}'.format(i+1))
        break
```
