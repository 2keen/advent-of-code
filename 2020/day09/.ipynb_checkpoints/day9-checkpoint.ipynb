```python
def validate(nums_list, preamble_length):
#     preamble = nums_list[:preamble_length]
    valid_list = []
    for i in range(len(nums_list)):
#         print(i,': ',end='')
#         print(i)
        if i < preamble_length:
#             print('preamble:',nums_list[i])
            continue
        else:
#             print('summing:',nums_list[i])
            sums = []
            for x in range(preamble_length):
                sum_index1 = i - preamble_length + x
                for y in range(1,preamble_length-x):
                    sum_index2 = sum_index1 + y
                    sum_val = nums_list[sum_index1] + nums_list[sum_index2]
                    sums.append(sum_val)
#                     print(x,y,sum_index1,sum_index2, nums_list[sum_index1], nums_list[sum_index2], sum_val)
            valid_list.append((i,nums_list[i],nums_list[i] in sums))
#             print(nums_list[i], nums_list[i] in set(sums))
#             print(set(sums))
    return valid_list
```

```python
def first_false(valid_list):
    for i,n,b in valid_list:
        if not b:
            return i,n
    return False
```
