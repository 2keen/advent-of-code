```python
import pandas as pd
```

```python
def double_digit(s):
    p = '0'
    for l in s:
        if l==p: return True
        p = l
    return False
```

```python
def ascending(s):
    p = '0'
    for l in s:
        if l<p: return False
        p=l
    return True
```

```python
pos = (265275, 781584)
df = pd.DataFrame({'possible':range(pos[0], pos[1]+1)})
df['in-range'] = df['possible'].between(pos[0],pos[1]+1)
df['double-digit'] = df['possible'].astype(str).apply(double_digit)
df['ascending'] = df['possible'].astype(str).apply(ascending)
df['all'] = df.all(axis=1)

df
```

```python
df[df['all']].count()
```

```python
df['all'].value_counts()
```

```python
#part 2
def triple_digit(s):
    p = '0'
    ct = 1
    for i,l in enumerate(s):
        if l==p:
            ct+=1
            if i+1<len(s):
                if s[i+1] != l and ct==2: return True
            elif ct==2: return True
                
        else:
            ct=1
        p = l
    return False
```

```python
df['triple-digit'] = df['possible'].astype(str).apply(triple_digit)
df = df.drop(['all'],axis=1)
df['all'] = df.all(axis=1)
df['all'].value_counts()
```
